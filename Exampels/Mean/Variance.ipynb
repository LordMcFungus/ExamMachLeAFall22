```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

# Mean / Variance

```python
N = 100_000_000
BINS = 200
```

```python
x = np.random.uniform(0,2,N)
plt.hist(x,bins=BINS)
plt.title("Uniform")
plt.show()
print(f"mean={np.mean(x)},var={np.var(x)}")
```

```python
x = np.random.normal(1,1,N)
plt.hist(x,bins=BINS)
plt.title("Gauss")
plt.show()
print(f"mean={np.mean(x)},var={np.var(x)}")
```

```python
x = np.random.poisson(1,N)
plt.hist(x,bins=BINS)
plt.title("Poisson")
plt.show()
print(f"mean={np.mean(x)},var={np.var(x)}")
```
